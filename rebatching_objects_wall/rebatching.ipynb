```python
import pandas as pd
import os
import shutil
```

```python
object_wall_df = pd.read_csv("../object_wall_hanging_report/object_wall_hanging_coding.csv")
```

```python
needs_rebatching = object_wall_df[object_wall_df["error"].notnull()].copy() 
needs_rebatching.loc[:, "filename"] = needs_rebatching["image_id"].str.split(".").str[0] + ".jpg"
needs_rebatching.drop_duplicates(subset="filename", inplace=True)
needs_rebatching
```

```python
images_dir = r"C:\Users\ethan_haque\Desktop\tmp\objects_wall_hanging"
```

```python
# get list of all file paths whose name matches filename
def get_file_path(filename):
    for root, dirs, files in os.walk(images_dir):
        for file in files:
            if file == filename:
                return os.path.join(root, file)
```

```python
needs_rebatching["file_path"] = needs_rebatching["filename"].apply(get_file_path)
needs_rebatching
```

```python
new_dir = r"C:\Users\ethan_haque\Desktop\tmp\objects_wall_hanging\rebatched"
os.makedirs(new_dir, exist_ok=True)
for file_path in needs_rebatching["file_path"]:
    shutil.copy(file_path, new_dir)
```
